### 1. Get weather data with an API 

*   Use https://nominatim.org/ to get the gps coordinates of all the cities (no subscription required) Documentation : https://nominatim.org/release-docs/develop/api/Search/



```python
import requests 
import pandas as pd
import json
```

```python
url = "https://nominatim.openstreetmap.org/search?q={}&country=<france>&format=json&limit=1"

for i in range(df.shape[0]) :
    r = requests.get(url.format(df.loc[i, 'cities']))
    rj = r.json()
    df.loc[i, "latitudes"] = rj[0]['lat']
    df.loc[i, "longitudes"] = rj[0]['lon']
```
